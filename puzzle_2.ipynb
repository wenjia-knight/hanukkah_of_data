```python
import pandas as pd

products = pd.read_csv("data/noahs-products.csv")
coffees = products['desc'].str.contains('Coffee')
bagels = products['desc'].str.contains('Bagel')
products = products[coffees | bagels]
products
```

```python
orders = pd.read_csv("data/noahs-orders.csv", parse_dates=["ordered", "shipped"])
order_items = pd.read_csv("data/noahs-orders_items.csv")
orders = orders.merge(order_items)
orders = orders[(orders["ordered"].dt.year == 2017) & ((orders["ordered"] - orders["shipped"]).dt.seconds <= 60)]
orders = orders.merge(products)
orders
```

```python
customers = pd.read_csv("data/noahs-customers.csv")
customers
```

```python
merged = pd.merge(customers, orders, left_on="customerid", right_on="customerid")
merged
```

```python
names = customers["name"].str.split(expand=True)
print(names)
customers["first_name"] = names[0]

if names[2].isnull:
    customers["last_name"] = names[1]
if names[2].isnull == False:
    customers["last_name"] = names[2]
customers
condition_1 = customers["first_name"].str[0] == 'J'
condition_2 = customers["last_name"].str[0] == 'D'
customers = customers[condition_1 & condition_2].reset_index()
customers
```

```python
new_customers = customers.merge(orders, how='left',on='customerid')
new_customers
```
