```python
import pandas as pd
df_products = pd.read_csv("data/noahs-products.csv")
df_products = df_products[df_products['desc'].str.contains("\)")]
split = df_products['desc'].str.split('\(', expand=True)

df_products['desc'] = split[0]
df_products['colour'] = split[1].str.replace(')',"")
df_products
```

```python
df_orders_items = pd.read_csv("data/noahs-orders_items.csv")
df_orders_items
```

```python
df_orders_items = df_orders_items.merge(df_products)
df_orders_items
```

```python
df_orders = pd.read_csv("data/noahs-orders.csv", parse_dates=['ordered', 'shipped'])


df_orders = df_orders.merge(df_orders_items)

df_orders
```

```python
df_emily = df_orders[df_orders["customerid"] == 8342]
df_emily
```

```python
desc = df_emily['desc'].tolist()
df_orders = df_orders[df_orders['desc'].isin(desc)]
df_orders
```

```python
colours = df_emily['colour'].tolist()
df_orders = df_orders[~df_orders['colour'].isin(colours)]
df_orders
```

```python
condition_1 = df_orders['ordered'].dt.date.isin(df_emily['ordered'].dt.date)

possible_orders = df_orders[condition_1]
```

```python
df_customers = pd.read_csv("data/noahs-customers.csv")
possible_orders = possible_orders.merge(df_customers)
possible_orders
```
