```python
import pandas as pd
df_orders = pd.read_csv("data/noahs-orders.csv")
df_orders
```

```python
df_products = pd.read_csv("data/noahs-products.csv")
df_products
```

```python
df_orders_items = pd.read_csv("data/noahs-orders_items.csv")
df_orders_items = df_orders_items.merge(df_products)
df_orders_items
```

```python
orders_costs = df_orders_items.groupby('orderid')[['unit_price', 'wholesale_cost']].sum()
new_df = orders_costs.reset_index()
new_df
```

```python
df_orders = df_orders.merge(new_df)
df_orders
df_orders['profit'] = (df_orders['total'] - df_orders['wholesale_cost']).round(2)
at_loss = df_orders[df_orders['profit'] < 0]

at_loss
```
