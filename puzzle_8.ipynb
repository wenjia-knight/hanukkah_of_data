```python
import pandas as pd
df_products = pd.read_csv("data/noahs-products.csv")
df_products = df_products[df_products['sku'].str.contains("COL")]
df_products
```

```python
df_orders_items = pd.read_csv("data/noahs-orders_items.csv")
df_orders_items
```

```python
df_orders_items = df_orders_items.merge(df_products)
df_orders_items
```

```python
df_orders = pd.read_csv("data/noahs-orders.csv")
df_orders
```

```python
df_orders_items = df_orders_items.merge(df_orders)
df_orders_items
```

```python
df_customers = pd.read_csv("data/noahs-customers.csv")
df_customers
```

```python
df_orders_items = df_orders_items.merge(df_customers)
df_orders_items
```
