```python
# create a dictionary to match each letter to a number
matching_numbers = {
    0: [""],
    1: [""],
    2: ["a", "b", "c"],
    3: ["d", "e", "f"],
    4: ["g", "h", "i"],
    5: ["j", "k", "l"],
    6: ["m", "n", "o"],
    7: ["p", "q", "r", "s"],
    8: ["t", "u", "v"],
    9: ["w", "x", "y", "z"],
}
```

```python
def name_to_num(name: str) -> str:
    name = name.lower()
    phone_number_list = []
    for letter in name:
        matching_num = [k for k, v in matching_numbers.items() if letter in v]
        phone_number_list.append(str(matching_num[0]))
    phone_number = "".join(phone_number_list)
    return phone_number
```

```python
df['name_to_number'] = df['last_name'].map(name_to_num)
df
```
